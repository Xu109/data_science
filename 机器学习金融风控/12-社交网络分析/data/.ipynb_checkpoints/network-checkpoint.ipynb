```python
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
```

```python
edge_list=pd.read_csv('stack_network_links.csv')
edge_list.head()
```

```python
G=nx.from_pandas_edgelist(edge_list)

plt.figure(figsize=(20,20))

nx.draw(G,with_labels=True,
        edge_color='grey',
        node_color='blue',
        node_size=10,
        pos=nx.spring_layout(G,k=0.2,iterations=50))
```

```python
nx.degree(G)
nx.degree_centrality(G)
nx.betweenness_centrality(G)
nx.closeness_centrality(G)
nx.katz_centrality(G)
```

```python
Sub_G=G.subgraph(list(nx.connected_components(G))[0])# 取出网络的最大子图
nx.average_clustering(Sub_G) #计算网络的聚类系数
nx.average_shortest_path_length(Sub_G) #计算网络的平均路径
```
